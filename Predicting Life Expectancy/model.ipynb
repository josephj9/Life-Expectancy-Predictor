```python
import pandas as pd
data = pd.read_csv("Life Expectancy Data.csv")
data
```

```python
data.corr(numeric_only="true")["Life expectancy "]

```

```python
data = data[["Country", "Year", "Status", "Life expectancy ", "Adult Mortality", "Alcohol", "percentage expenditure", 
             " BMI ", "Polio", "Diphtheria ", " HIV/AIDS", "GDP",  " thinness  1-19 years", " thinness 5-9 years", 
             "Income composition of resources", "Schooling"]]
data
```

```python
data.plot.hist(y="Life expectancy ")
```

```python
data[data.isnull().any(axis=1)]
```

```python
data = data.dropna()
data
```

```python
data["isDeveloping"] = (data["Status"] == "Developing").astype(int)
data
```

```python
from sklearn.model_selection import train_test_split  
X=data[["Adult Mortality", "Alcohol", "percentage expenditure", 
             " BMI ", "Polio", "Diphtheria ", " HIV/AIDS", "GDP",  " thinness  1-19 years", " thinness 5-9 years", 
             "Income composition of resources", "Schooling", "isDeveloping"]]
y=data["Life expectancy "]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, )  
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
y_test
```

```python
reg.score(X_test,y_test)
```
